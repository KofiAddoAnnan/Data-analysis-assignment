QUESTION 6

An expensive wine lover is travelling to the US. He is going to tour different provinces or states.

He doesnt want wine below 250 dollars.

Give him a list of pronvinces (states) he can get wine according to his budget.

This would help him with informed decisions and not waste time visiting all wine shops whilst touring

```python
wine.head()
```

```python
united_states = wine[wine['country'] == 'US']
```

```python
united_states.head()
```

```python
len(united_states)
```

```python
wine_250_or_above = united_states[united_states['price'] >= 250]
```

```python
wine_250_or_above.head()
```

```python
len(wine_250_or_above)
```

```python
wine_250_or_above['province'].unique()
```

According to the above analysis, the provinces he would get wine according to his budget will be California and Michigan
